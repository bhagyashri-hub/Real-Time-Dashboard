```python


# Step 1: Imports
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
import random
import warnings

# Optional: suppress warnings for cleaner notebook
warnings.filterwarnings("ignore")

# Step 2: Create initial sample data
data = {
    "time": ["09:00", "10:00", "11:00", "12:00", "13:00"],
    "sales": [100, 150, 130, 170, 160],
    "traffic": [200, 250, 300, 270, 320]
}
df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)

# Step 3: Initialize Dash app
app = Dash(__name__)

# Step 4: Layout
app.layout = html.Div(style={'backgroundColor': '#1e1e2f', 'color': 'white', 'padding': '20px'}, children=[
    html.H1(" Real-Time Dashboard", style={'textAlign': 'center', 'marginBottom': '30px'}),
    
    # KPI cards
    html.Div([
        html.Div(id='sales-kpi', style={
            'width': '32%', 'display': 'inline-block',
            'padding': '20px', 'backgroundColor': '#ff4c4c',
            'color':'white', 'textAlign':'center', 'borderRadius':'10px', 'marginRight':'2%'
        }),
        html.Div(id='traffic-kpi', style={
            'width': '32%', 'display': 'inline-block',
            'padding': '20px', 'backgroundColor': '#4caf50',
            'color':'white', 'textAlign':'center', 'borderRadius':'10px', 'marginRight':'2%'
        }),
        html.Div(id='total-kpi', style={
            'width': '32%', 'display': 'inline-block',
            'padding': '20px', 'backgroundColor': '#2196f3',
            'color':'white', 'textAlign':'center', 'borderRadius':'10px'
        })
    ], style={'marginBottom':'30px'}),
    
    # Charts
    html.Div([
        dcc.Graph(id='line-graph', style={'display':'inline-block', 'width':'100%'}),
        dcc.Graph(id='bar-graph', style={'display':'inline-block', 'width':'48%', 'marginRight':'4%'}),
        dcc.Graph(id='pie-graph', style={'display':'inline-block', 'width':'48%'})
    ]),
    
    # Interval for updates
    dcc.Interval(
        id='interval-component',
        interval=5*1000,  # update every 5 seconds
        n_intervals=0
    )
])

# Step 5: Callback for updates
@app.callback(
    [Output('line-graph', 'figure'),
     Output('bar-graph', 'figure'),
     Output('pie-graph', 'figure'),
     Output('sales-kpi', 'children'),
     Output('traffic-kpi', 'children'),
     Output('total-kpi', 'children')],
    [Input('interval-component', 'n_intervals')]
)
def update_dashboard(n_intervals):
    # Load data
    df = pd.read_csv("data.csv")
    
    # Simulate new row
    last_hour = int(df['time'].iloc[-1].split(":")[0])
    new_time = f"{last_hour + 1}:00"
    new_sales = max(0, df['sales'].iloc[-1] + random.randint(-20, 20))
    new_traffic = max(0, df['traffic'].iloc[-1] + random.randint(-30, 30))
    
    df = pd.concat([df, pd.DataFrame([{"time": new_time, "sales": new_sales, "traffic": new_traffic}])], ignore_index=True)
    df = df.tail(10)  # keep last 10 rows
    df.to_csv("data.csv", index=False)
    
    # Line chart
    line_fig = px.line(df, x='time', y=['sales', 'traffic'], markers=True)
    line_fig.update_layout(
        title="Sales & Traffic Over Time",
        xaxis_title="Time",
        yaxis_title="Value",
        plot_bgcolor='#1e1e2f',
        paper_bgcolor='#1e1e2f',
        font_color='white'
    )
    
    # Bar chart
    bar_fig = px.bar(df, x='time', y=['sales', 'traffic'], barmode='group')
    bar_fig.update_layout(
        title="Sales vs Traffic (Bar Chart)",
        plot_bgcolor='#1e1e2f',
        paper_bgcolor='#1e1e2f',
        font_color='white'
    )
    
    # Pie chart
    pie_fig = px.pie(df.tail(1), names=['sales','traffic'], values=[df['sales'].iloc[-1], df['traffic'].iloc[-1]])
    pie_fig.update_layout(
        title="Proportion of Sales vs Traffic",
        plot_bgcolor='#1e1e2f',
        paper_bgcolor='#1e1e2f',
        font_color='white'
    )
    
    # KPI cards
    sales_kpi = [html.H3("Latest Sales"), html.H1(f"{df['sales'].iloc[-1]}")]
    traffic_kpi = [html.H3("Latest Traffic"), html.H1(f"{df['traffic'].iloc[-1]}")]
    total_kpi = [html.H3("Total (Sales+Traffic)"), html.H1(f"{df['sales'].iloc[-1] + df['traffic'].iloc[-1]}")]
    
    return line_fig, bar_fig, pie_fig, sales_kpi, traffic_kpi, total_kpi

# Step 6: Run inline in Jupyter
app.run(mode='inline', debug=True)
```
